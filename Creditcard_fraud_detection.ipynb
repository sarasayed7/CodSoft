The objective of this project is to develop a machine learning model capable of accurately detecting fraudulent credit card transactions. ### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/creditcard.csv')
```

### Splitting data into Training and Testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

Splitting data into 80% as training data and 20% as testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

### Scaling dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

Note:

Make sure to fit and transform training data.

And only transform testing data to avoid data leakage.

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

### Import Logistic Regression Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(scaled_X_train, y_train)
```

### Importing metrics

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred = model.predict(scaled_X_test)
```

Confusion matrix

```python
cm = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['Is Fraud', 'Is Normal'], columns=['Predicted Fraud', 'Predicted Normal'])
confusion
```

Displaying Confusion Matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm_plot = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

Printing Classification report

```python
print(classification_report(y_test, y_pred))
```

Displaying Precision Recall Curve

```python
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay
```

```python
precision, recall, _ = precision_recall_curve(y_test, y_pred)

disp = PrecisionRecallDisplay(precision=precision, recall=recall)

disp.plot()
plt.show()
```

Displaying ROC curve

```python
from sklearn.metrics import roc_curve, RocCurveDisplay
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred)
disp = RocCurveDisplay(fpr=fpr, tpr=tpr)

disp.plot()
plt.show()
```
